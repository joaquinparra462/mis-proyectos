# OneHotEncoder

```python
import pandas as pd
df = pd.read_csv("https://github.com/codebasics/py/raw/refs/heads/master/ML/5_one_hot_encoding/homeprices.csv")
df
```

```python
#Convierte variables categorias en binarias
dummies = pd.get_dummies(df.town)
```

```python
concat = pd.concat([df,dummies] , axis = 'columns')
concat
```

```python
final = concat.drop(["town", "west windsor"], axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
X = final.drop("price", axis='columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X,y) #se entrena el modelo con fit usando "linear regression"
```

```python
model.predict([[2800,0,1]])  #valor de price, monroe township, robinsville
```

```python
model.score(X,y)
```

# EJERCICIO

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/codebasics/py/raw/refs/heads/master/ML/5_one_hot_encoding/Exercise/carprices.csv")
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies 
```

```python
concat = pd.concat([df,dummies], axis= 'columns')
concat
```

```python
# sacamos una columna de datos para que ML no se ajuste demasiado a estos datos
final = concat.drop(['Car Model', 'Mercedez Benz C class'], axis= 'columns')
final
```

```python
X = final.drop(['Sell Price($)'], axis='columns') #se saca la data para usarla de entrenamiento
X
```

```python
y = final['Sell Price($)'] #data de entrenamiento
y
```

```python
#utilizando LabelEncoder para convertir datos categoricos a numericos
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

dfle = df
dfle['Car Model']= le.fit_transform(dfle['Car Model'])
dfle
```

```python
#se reasigna X despues de LabelEncoder SIN CONTAR con la data de entrenamiento
X = dfle[['Car Model', 'Mileage', 'Age(yrs)']].values
X
```

```python
y = dfle['Sell Price($)']
y
```

```python
from sklearn.preprocessing import OneHotEncoder
#debido a que queremos utilizar la columna[0], se utiliza ColumnTransformer ya que las nuevas versiones de notebook
# no usan categorical_features

from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[('onehot', OneHotEncoder(), [0])],
    remainder= 'passthrough'
)
```

```python
X = ct.fit_transform(X)
X
```

```python
X = np.delete(X,0, axis =1)
X
```

```python
#Se entrena el modelo
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
#Ahora se hace la prediccion que se solicita
#prediciendo un mercedez benz que tiene 4años y kilometraje de 45000
model.predict([[0,1,45000,4]])
```

```python
#Price of BMW X5 that is 7 yr old with mileage 86000
model.predict([[1,0,86000,7]])
```

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv("https://github.com/codebasics/py/raw/refs/heads/master/ML/2_linear_reg_multivariate/homeprices.csv")
```

```python
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area", "bedrooms", "age"]], df.price)  #variables independientes en la x, dependientes/target en la y
```

```python
reg.predict([[3000,3,40]])
```

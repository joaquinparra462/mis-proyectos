# Training and Test Data


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/codebasics/py/raw/refs/heads/master/ML/6_train_test_split/carprices.csv")
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("Milage")
plt.ylabel("Sell Price")
plt.scatter(df['Mileage'],df['Sell Price($)'])
```

```python
plt.xlabel("Age")
plt.ylabel("Sell Price")
plt.scatter(df['Age(yrs)'],df['Sell Price($)'])
```

```python
X = df[['Mileage', 'Age(yrs)']]
y = df['Sell Price($)']
X,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2) # train:80 / test:20
#si se quiere se genere siempre los mismo datos, se especifica el random_state
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
```

```python
#entrenando al modelod e machine learning
clf.fit(X_train, y_train)
```

```python
#se hace la prediccion con los test utilizando el modelo
clf.predict(X_test)
```

```python
#se compara el modelo usando score para comprobar la precision. Se usan ambos test.
clf.score(X_test, y_test)
```

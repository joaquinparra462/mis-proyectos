```python
bios_new['height_category'] = bios_new['height_cm'].apply(lambda x:'Short' if x < 165 else('Average' if x < 185 else 'Tall'))
#lambda seria una funcion anonima que toma un parametro x
#crea una categoria y la clasifica como bajo, promedio o alto.
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'LightWeight'
    elif row['height_cm']< 175 and row['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'HeavyWeight'

bios_new['Category'] = bios.apply(categorize_athlete, axis=1) #1 es fila, 0 son columnas
```

# Fusionar y Concatenar Data

```python
noc = pd.read_csv("./data/noc_regions.csv")
```

```python
#esto mostrara un NOC_x y un NOC_y, esto se debe a que cuando se fusionan DataFrames en pandas y ambos tienen
#una columna con el mismo nombre, pandas las separa automaticamente con los prefixos 'x' e 'y'.
#el prefijo 'x' refiere al DataFrame de la izquierda, mientras que la 'y' refiere al de la derecha.
#en este x=bios , y=noc
bios_new = pd.merge(bios,noc, left_on='born_country', right_on='NOC', how='inner') #junta el pais de nacimiento con el pais representado
```

```python
bios_new.rename(columns={'region': 'born_country_full'}, inplace=True)
```

```python
bios_new[bios_new['NOC_x'] != bios_new['born_country_full'] ]
#muestra solo los paises que NO coinciden con el pais de nacimiento
#en este caso serian personas que nacieron en un pais pero despues representaron a otro en las olimpiadas
#NOC_x, NOC_y pueden ser distintos ya que al fusionar ambos DF, uno puede tener la data actualizada y otro no.
```

# Manejando valores nulos

```python
coffe_new = coffe_new.fillna(coffe['Units Sold'].mean()) #se rellenan los valores nulos o vacios con el promedio de la columna 'Units Sold'
```

```python
coffe_new['Units Sold'] = coffe_new['Units Sold'].interpolate().bfill().ffill()
#bfill = backward fill
#ffill = forward fill
#los metodos se ejecutan uno tras del otro, primero interpolando, luego rellena desde atras, y termina rellenando hacia adelante,
#asegurandose de que todos los valores nulos en Units Sold se completen
```

# Aggregating Data

```python
bios[bios['born_country'] == 'CHI']['born_region'].value_counts()
#CHILE EN LOS OLIMPICOS, VAMOOOOOH !!!!
```

```python
coffe_new.groupby(['Coffee Type', 'Day']).agg({'Units Sold': 'sum', 'price': 'mean'})
#agrupa los datos de coffee por tipo de cafe y dia, calculando para Units sold la suma y price el promedio
```

```python
pivot = coffe_new.pivot(columns = 'Coffee Type', index= 'Day', values='revenue')
#pivot transforma filas en columnas. agrega data basada en indices especificos y agrupacion de columnas
#tomas los valores unicos de especificados de 'columns' y las convierte en nuevos encabezados de columnas
#usa los valores del parametro 'index' como filas nuevas
#rellena las celdas intersectadas con valores de el paramtro 'values'

#ej
#Original "long" format:
#Day    Coffee Type    Revenue
#Mon    Espresso      100
#Mon    Latte         150
#Tue    Espresso      120
#Tue    Latte         180

#After pivot - "wide" format:
#Day    Espresso    Latte
#Mon    100         150
#Tue    120         180
```

# Funcionalidad Avanzada

```python
#algunas fucnciones: .shift(), .rank(), .rolling(), .cumsum()
coffe_new['yesterday_revenue'] = coffe_new['revenue'].shift(2)
#se usa shift(2) porque el dataset va de 2 en 2 debido a los tipos de cafe
```

```python
coffe_new['pct_change'] = coffe_new['revenue'] / coffe_new['yesterday_revenue'] * 100
```

```python
coffe_new
```

```python
bios_new['height_rank'] = bios_new['height_cm'].rank(ascending=False) #compara la estatura de todos y les da un rating desde el mas alto al mas bajo
```

```python
bios_new.sort_values(['height_rank']) 
```

```python
coffe_new.info()
```

```python
coffe_new['cumulative_revenue'] = coffe_new['revenue'].cumsum()
#suma todos los ingresos y los va acumulando sucesivamente.

coffe_new
```

```python
latte = coffe_new[coffe_new['Coffee Type'] == 'Latte' ].copy()
latte['3day'] = latte['Units Sold'].rolling(3).sum()

#ej:
#**Example with `rolling(3).sum()`:**
#- If you have data: [10, 20, 30, 40, 50]
#- Position 1: NaN (not enough previous values)
#- Position 2: NaN (still not enough previous values)  
#- Position 3: 10 + 20 + 30 = 60
#- Position 4: 20 + 30 + 40 = 90
#- Position 5: 30 + 40 + 50 = 120

#The rolling method is commonly used for:
#- Moving averages
#- Trend analysis
#- Smoothing noisy data
#- Time series analysis

latte
```
